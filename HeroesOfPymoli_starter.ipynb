```python
#Purchasing Analysis(TOTAL)
#Basic Calculations
unique_items = len(purchase_data_df["Item ID"].unique())
average_price = purchase_data_df["Price"].mean()
number_of_purchases = purchase_data_df["Price"].count()
total_revenue = purchase_data_df["Price"].sum()


#Summary data frame to hold results
summary_table = pd.DataFrame({"Number of Unique Items": unique_items,
                               "Average Purchase Price": [average_price],
                               "Number of Purchases": [number_of_purchases],
                               "Total Revenue": [total_revenue]})
#Display summary
summary_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Purchasing Analysis(Age)
#Bin the purchase_data frame by age
age_df = pd.DataFrame(purchase_data_df)
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">40"]
bins_for_ages = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 9999999]
age_demo_grp = age_df.groupby(pd.cut(age_df["Age"], bins_for_ages, labels=age_groups))


#Create a summary data frame to hold the results
age_demo_df = pd.DataFrame({"Purchase Count":age_demo_grp["Price"].count(), "Average Purchase Price":age_demo_grp["Price"].mean(),
                            "Total Purchase Value":age_demo_grp["Price"].sum(),
                            "Ave Total Purchase per Person":age_demo_grp["Price"].sum()/age_demo_grp["SN"].nunique()})

#Summary data 
age_demo_df["Average Purchase Price"] = age_demo_df["Average Purchase Price"]
age_demo_df["Total Purchase Value"] = age_demo_df["Total Purchase Value"]
age_demo_df["Ave Total Purchase per Person"] = age_demo_df["Ave Total Purchase per Person"]

#Display the summary data frame
age_demo_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Most Popular
#Group by Item ID and Item Name.
popular_items_df = pd.DataFrame(purchase_data_df)
popular_items_group = popular_items_df.groupby(['Item ID','Item Name'])

#Perform calculations to obtain purchase count, item price, and total purchase value
#Create a summary data frame to hold the results
popular_items_df = pd.DataFrame({"Purchase Count":popular_items_group["Price"].count(), 
                            "Item Price":popular_items_group["Price"].mean(),
                            "Total Purchase Value":popular_items_group["Price"].sum()})


#Sort the purchase count column in descending order
popular_items_df = popular_items_df.sort_values("Purchase Count", ascending=False)
popular_items_df["Item Price"] = popular_items_df["Item Price"]
popular_items_df["Total Purchase Value"] = popular_items_df["Total Purchase Value"]

#Display a preview of the summary data frame
popular_items_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Most Profitable Items

#Identify the 5 most profitable items by total purchase value, then list (in a table):

mostprofit_df = pd.DataFrame(purchase_data_df)
mostprofit_group = mostprofit_df.groupby(['Item ID','Item Name'])
mostprofit_df = pd.DataFrame({"Purchase Count":mostprofit_group["Price"].count(), 
                              "Item Price":mostprofit_group["Price"].mean(),
                              "Total Purchase Value":mostprofit_group["Price"].sum()})

#Perform calculations to obtain purchase count, item price, and total purchase value
#Sort the purchase item column in descending order
mostprofit_df = mostprofit_df.sort_values("Total Purchase Value", ascending=False)
mostprofit_df["Item Price"] = mostprofit_df["Item Price"]
mostprofit_df["Total Purchase Value"] = mostprofit_df["Total Purchase Value"]

mostprofit_df.head()
```
